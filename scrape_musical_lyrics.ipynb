```python
#get soup for a page about a musical, extract title and description, get all song lyric links and lyrics
def songs_from_a_musical(url):
    url = url
    response = requests.get(url)
    response.status_code # status code = 200 => OK
    page = response.text
    soup = BeautifulSoup(page, "lxml")
    
    #basic info about show
    title = soup.find("h2").text
    title = title.replace(' synopsis','')
    print(title)
    description = soup.find(id="page").text
    #date = re.findall(re.compile(r'Release date: .*([1-3][0-9]{3})'), description)
    
    #get song lyric urls
    urls_on_page = soup.find_all("a")
    indices = [i for i, s in enumerate(urls_on_page) if 'www.allmusicals.com/lyrics/' in str(s)]
    lyric_links =[urls_on_page[i] for i in indices]
    lyric_urls = [re.search(r'href=(.*?)>', str(i)).group(1) for i in lyric_links]
    lyric_urls = [i.strip('\"') for i in lyric_urls]
    all_lyrics = collect_lyrics(lyric_urls)
    
    this_musical_text = pd.DataFrame(columns=['titles','descriptions','song_lyrics'])
    for j in range(len(all_lyrics)):
        d = {'titles':title, 'descriptions':description, 'song_lyrics':all_lyrics[j]}
        this_musical_text = this_musical_text.append(d, ignore_index = True)
    return this_musical_text
```

```python
#loop through each of these and scrape lyrics from those pages
def collect_lyrics(lyric_urls):
    song_lyric = []
    for i in lyric_urls:
        response = requests.get(i)
        response.status_code # status code = 200 => OK
        page = response.text
        soup = BeautifulSoup(page, "lxml")
        try:
            song_lyric.append(soup.find(id="page").text)
        except AttributeError:
            continue
    return song_lyric
```
